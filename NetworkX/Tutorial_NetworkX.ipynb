# Drawing Weighted Graph

```python
import networkx as nx
import matplotlib.pyplot as plt
```

- Graph init

```python
G = nx.Graph()
```

- Adding edge

```python
G.add_edge('a', 'b', weight=0.63)
G.add_edge('a', 'c', weight=0.28)
G.add_edge('c', 'd', weight=0.11)
G.add_edge('c', 'e', weight=0.72)
G.add_edge('c', 'f', weight=0.91)
G.add_edge('a', 'd', weight=0.33)
G.add_edge('d', 'h', weight=0.38)
G.add_edge('h', 'a', weight=0.88)
```

- quarter

```python
q1 = [(u, v) for (u, v, d) in G.edges(data=True) if d['weight'] <= 0.25]
q2 = [(u, v) for (u, v, d) in G.edges(data=True) if d['weight'] > 0.25 and d['weight'] <= 0.5]
q3 = [(u, v) for (u, v, d) in G.edges(data=True) if d['weight'] > 0.5 and d['weight'] <= 0.75]
q4 = [(u, v) for (u, v, d) in G.edges(data=True) if d['weight'] > 0.75]
```

- Positioning all nodes

```python
pos = nx.spring_layout(G)
```

```python
pos
```

- Drawing

```python
nx.draw_networkx_nodes(G, pos, node_size=700)
nx.draw_networkx_edges(G, pos, edgelist=q1, alpha=0.25, width=5)
nx.draw_networkx_edges(G, pos, edgelist=q2, alpha=0.5, width=5)
nx.draw_networkx_edges(G, pos, edgelist=q3, alpha=0.75, width=5)
nx.draw_networkx_edges(G, pos, edgelist=q4, alpha=1.0, width=5)
nx.draw_networkx_labels(G, pos, font_size=20, font_family='sans-serif')
plt.axis('off')
plt.show()
```

# Graph Operation

- subgraph(G, nbunch): induced subgraph view of G on nodes in nbunch
- union(G1,G2): graph union
- disjoint_union(G1,G2): graph union assuming all nodes are different
- cartesian_product(G1,G2): return Cartesian product graph
- compose(G1,G2): combine graphs identifying nodes common to both
- complement(G): graph complement
- create_empty_copy(G): return an empty copy of the same graph class
- to_undirected(G): return an undirected representation of G
- to_directed(G): return a directed representation of G

```python
petersen = nx.petersen_graph()
```

```python
nx.draw(petersen, with_labels=True, font_weight='bold')
```

```python
K_3_5 = nx.complete_bipartite_graph(3, 5)
```

```python
nx.draw(K_3_5, with_labels=True, font_weight='bold')
```

```python
cartesian_product_graph = nx.cartesian_product(petersen, K_3_5)
```

```python
nx.draw(cartesian_product_graph, with_labels=True, font_weight='bold')
```
